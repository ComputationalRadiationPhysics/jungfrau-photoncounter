```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from os import listdir
%matplotlib notebook
```

```python
cuda1_files = [f for f in listdir("../build_cuda_1/") if "clustercount" in f and "_5_" in f]
clustercount = [int(f[12:].split("_")[0]) for f in cuda1_files]
cuda1_means = []
cuda1_stddev = []
```

```python
for i in range(len(clustercount)):
    d = np.loadtxt("../build_cuda_1/" + cuda1_files[i])/1000000000.
    cuda1_means += [np.mean(d[1:])]
    cuda1_stddev += [np.std(d[1:])]
```

```python
plt.scatter(clustercount, cuda1_means)
plt.xlabel("Amount of Cluster Centers [in % of Total Pixels]")
plt.ylabel("Runtime [in s]")
plt.title("Runtime with Levels of Cluster Usage")
plt.show()
```

```python
cuda1_means = np.array(cuda1_means)
clustercount = np.array(clustercount)
min_elem = cuda1_means[np.where(clustercount == 0)][0]
plt.scatter(clustercount, cuda1_means/min_elem)
plt.xlabel("Amount of Cluster Centers [in % of Total Pixels]")
plt.ylabel("Speedup")
plt.title("Runtime with Levels of Cluster Usage")
plt.show()
```
